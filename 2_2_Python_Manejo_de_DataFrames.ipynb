# Limpieza de Datos


---



La limpieza de datos implica identificar y rectificar errores, inconsistencias y valores faltantes dentro de nuestro conjunto de datos. Es necesario preparar nuestros datos antes de procesarlos.

Los modelos de aprendizaje automático no se pueden entrenar con conjuntos de datos a los que les faltan valores "NAN", ya que pueden alterar el resultado final durante el análisis.

```python
#Eliminar las filas donde haya valores faltantes "NAN"
DB = DB.dropna()

#El tamaño del DataFrame ha cambiado
DB.shape
```

Los valores duplicados pueden alterar el análisis de resultados, pueden producirse valores atípicos.

```python
#Eliminar las filas duplicadas
DB = DB.drop_duplicates()

#El tamaño del DataFrame ha cambiado
DB.shape
```

# Edición del DataFrame


---



En ocasiones puede ser que necesitemos renombrar las columnas del DataFrame, esto podemos lograrlo aplicando la siguiente función a la variable `DB.rename` seguido de un nuevo diccionario que contenga a las columnas existentes asignadas con su nuevo nombre.

```python
#Renombrar columnas
DB.rename(columns = {'standard_value':'IC50', "standard_units": "Units"}, inplace = True)

#Exploramos el DataFrame con las columnas editadas
DB.head()
```

Como podemos notar, tenemos filas que tienen asignados varios valores separados por comas en una única celda en las tres columnas, sin embargo, con la función `.split` podemos hacer que tales valores se separen y resulten cada uno asignado a una fila diferente. Veamos el ejemplo para separar los valores de la columna `Units`, generaremos una nueva columna llamada `New_Units` con los valores separados:

```python
# Definir una función para separar los valores y crear nuevas filas
def separar_valores(row):
    valores = row['Units'].split(',')  # Cambia 'columna_con_valores' por el nombre de la columna que contiene los valores separados por comas
    return pd.Series(valores)

# Aplicar la función a la columna específica y generar nuevas filas
nuevo_df = DB.apply(separar_valores, axis=1).stack().reset_index(level=1, drop=True)
nuevo_df.name = 'New_Units'  # Cambia 'nueva_columna' por el nombre que quieras darle a la nueva columna

# Eliminar la columna original y unir el DataFrame original con el nuevo DataFrame
df = DB.drop('Units', axis=1).join(nuevo_df)
```

```python
#Explorar el nuevo DataFrame
df.head()
```

Para este ejemplo, únicamente requerimos unidades de nM o de uM, por lo tanto, podemos filtrar los datos mediante la columna `New_Units` de tal forma que las filas que contengan cualquier otra unidad se eliminen.

```python
# Filtrar filas con valores diferentes a "nM" o "uM" en la columna "New_Units"
df = df[df['New_Units'].isin(['nM', 'uM'])]
df.head()
```

```python
#El tamaño del DataFrame ha cambiado
df.shape
```

Ahora que ya conocemos cómo separar los datos separados por comas en una misma celda, es necesario aplicar la misma función para la columna `IC50`, guardaremos los datos en una nueva columna llamada `New_IC50`:

```python
# Definir una función para separar los valores y crear nuevas filas
def separar_valores1(row):
    valores1 = row['IC50'].split(',')  # Cambia 'columna_con_valores' por el nombre de la columna que contiene los valores separados por comas
    return pd.Series(valores1)

# Aplicar la función a la columna específica y generar nuevas filas
nuevo_df1 = df.apply(separar_valores1, axis=1).stack().reset_index(level=1, drop=True)
nuevo_df1.name = 'New_IC50'  # Cambia 'nueva_columna' por el nombre que quieras darle a la nueva columna

# Eliminar la columna original y unir el DataFrame original con el nuevo DataFrame
df1 = df.drop('IC50', axis=1).join(nuevo_df1)
df1.head()
```

Para efectos del ejemplo conservaremos únicamente el primer valor por cada blanco, por lo tanto, eliminamos los blancos duplicados:

```python
df1 = df1.drop_duplicates(subset='ID', keep='first')
df1.head()
```

Podemos realizar una limpieza final.

```python
#Eliminar las filas donde haya valores faltantes "NAN"
df1 = df1.dropna()
#df1 = df1.dropna(subset=['New_IC50'], inplace = True)
#El tamaño del DataFrame ha cambiado
df1.shape
```

Podemos guardar nuestro DataFrame como un archivo `.csv` aplicando la siguiente función:

```python
# Guardar el DataFrame resultante en un nuevo archivo CSV
df1.to_csv('CHEMBL_IC50_CURED.csv', index=False)
```

Dado que el cálculo del pIC50 es diferente según las unidades de medida que correspondan al valor, separaremos los datos cuyas unidades sean nM en un DataFrame 1 (df_1) y los datos cuyas unidades sean uM en otro DataFrame 2 (df_2)


```python
#Separar DataFrames
df_1 = df[df["New_Units"] == "nM"]
df_2 = df[df["New_Units"] == "uM"]
```

```python
#Exploramos el DataFrame 1
df_1.head()
```

```python
df_1 = df_1.copy()                                    # Creaar una copia del DataFrame
df_1['New_IC50'] = df_1['New_IC50'].astype(float)
df_1
```

Para calcular el pIC50 en unidades de nM, el orden de magnitud es de -9 en la fórmula, quedando:

```python
df_1["pIC50"] = df_1.apply(lambda x: convert_ic50_to_pic50(x.New_IC50), axis=1)
# We will sort the dataframe considering pIC50
df_1.sort_values(by="pIC50", ascending=False, inplace=True)
# Reset index
df_1.reset_index(drop=True, inplace=True)
df_1.head()
```

Ahora aplicamos los mismos pasos para el DataFrame 2

```python
#Exploramos el DataFrame 2
df_2.head()
```

```python
df_2 = df_2.copy()                                    # Creaar una copia del DataFrame
df_2['New_IC50'] = df_2['New_IC50'].astype(float)
df_2
```

Para calcular el pIC50 en unidades de uM, el orden de magnitud es de -6 en la fórmula, quedando:

```python
def convert_ic50_to_pic50(IC50_value):
    if IC50_value <= 0:
        return 0
    else:
        pIC50_value = 6 - math.log10(IC50_value)
        return pIC50_value
```

```python
df_2["pIC50"] = df_2.apply(lambda x: convert_ic50_to_pic50(x.New_IC50), axis=1)
# We will sort the dataframe considering pIC50
df_2.sort_values(by="pIC50", ascending=False, inplace=True)
# Reset index
df_2.reset_index(drop=True, inplace=True)
df_2.head()
```

Concatenamos ambos DataFrames 1 y 2 para obtener un único DataFrame con todos los valores calculados de pIC50:

```python
#Unir DataFrames
df_all = pd.concat([df_1, df_2])
df_all.head()
```

```python
#Exploramos nuestro DataFrame final
print(df_all)
```

```python
#Guardar archivo en formato csv
df_all.to_csv('pIC50_CHEMBL.csv', index=False)
```

**Análisis del pIC50**

Para poder determinar si un compuesto es "Activo" o "Inactivo" podemos agregar una nueva columna al DataFrame que nos indique respecto al pIC50 si el compuesto lo es o no, bajo la condición de que: "Un compuesto será activo si su pIC50 es mayor o igual a 5". El resto de los compuestos que no cumplan tal condición se etiquetan como inactivos.

Label compounds with pIC50 >= 5 as "active":



```python
#Nombrar los compuestos con pIC50 mayor o igual a 5 como "Active"
df_all["Activity_Label"] = np.where(df_all['pIC50']>=5, 'Active', 'Inactive')
df_all
```

Podemos realizar ahora la estadística por datos agrupados en la clasificación "Activo/Inactivo". Para agrupar datos se utiliza la función `.groupby` y aplicamos posteriormente la función `.describe`:

Group data by activity label:

```python
#Agrupar datos según la etiqueta de actividad "Activity_Label" que implica si el compuesto es "Activo/Inactivo"
group= df_all.groupby('Activity_Label').describe()['pIC50']
group
```
