```python
!pip install rdkit
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import PandasTools
from rdkit.Chem import Descriptors
df = pd.read_csv("/content/Antibacterial Drugs from 01_01_1981 to 12_31_2010 - Hoja 1 (1).csv")
PandasTools.AddMoleculeColumnToFrame(df, "Smile")
df["MW"] = df["ROMol"].apply(Descriptors.ExactMolWt)
df["HBA"] = df["ROMol"].apply(Descriptors.NumHAcceptors)
df["HBD"] = df["ROMol"].apply(Descriptors.NumHDonors)
df["logP"] = df["ROMol"].apply(Descriptors.MolLogP)
df["TPSA"] = df["ROMol"].apply(Descriptors.TPSA)
df["RotBonds"] = df["ROMol"].apply(Descriptors.NumRotatableBonds)
lipinski_filter = (
    (df["MW"] < 500) &
    (df["HBD"] <= 5) &
    (df["HBA"] <= 10) &
    (df["logP"] <= 5)
)

veber_filter = (
    (df["RotBonds"] <= 10) &
    (df["TPSA"] <= 140)
)

filtered_df = df[lipinski_filter & veber_filter]

filtered_df
```
